## Load Cleaned Dataset

The cleaned dataset produced in the previous notebook is loaded. Feature engineering is performed on this dataset to create additional variables that may improve the predictive performance of the machine learning model.

## Total Living Area

Living space is one of the strongest predictors of house price.

A new feature, `TotalLivingArea`, combines above-ground living area (`GrLivArea`) with basement area (`TotalBsmtSF`) to better represent the total usable living space available within a property.

```python
df["TotalLivingArea"] = df["GrLivArea"] + df["TotalBsmtSF"]

df[["GrLivArea", "TotalBsmtSF", "TotalLivingArea"]].head()
```

These engineered variables combine related property information into features that may provide useful predictive information during model development.

## House Age

Property age can influence market value.

A new feature, `HouseAge`, is created by calculating the difference between the year a property was remodelled and the year it was originally built.

```python
df["HouseAge"] = df["YearRemodAdd"] - df["YearBuilt"]

df[["YearBuilt", "YearRemodAdd", "HouseAge"]].head()
```

This feature captures how long it took before a property was renovated or improved, which may provide useful information for predicting sale prices.

Although this feature is named `HouseAge`, it represents the number of years between original construction and the most recent recorded remodelling year, rather than the age of the property at the time of sale.

## Total Porch Area

Outdoor living space may contribute to the value of a property.

A new feature is created by combining the sizes of open porches, enclosed porches and wooden decks into a single measurement.

```python
df["TotalPorchSF"] = df["OpenPorchSF"] + df["EnclosedPorch"] + df["WoodDeckSF"]

df[["OpenPorchSF", "EnclosedPorch", "WoodDeckSF", "TotalPorchSF"]].head()
```

Combining these related variables simplifies the dataset while retaining information about outdoor living space.

## Garage Feature

Rather than using only garage size, an additional binary feature is created to indicate whether a property has a garage.

```python
df["HasGarage"] = (df["GarageArea"] > 0).astype(int)

df[["GarageArea", "HasGarage"]].head()
```

Binary features are often easier for machine learning models to interpret than continuous variables alone.

## Second Floor Feature

A binary feature is created to indicate whether a property has a second floor.

```python
df["HasSecondFloor"] = (df["2ndFlrSF"] > 0).astype(int)

df[["2ndFlrSF", "HasSecondFloor"]].head()
```

This feature distinguishes single-storey and multi-storey properties while reducing the complexity of the original variable.
